```python
import os
import cv2
```

```python
def read_images_from_folder(folder_path):
    images = []
    for filename in sorted(os.listdir(folder_path)):
        img_path = os.path.join(folder_path, filename)
        if os.path.isfile(img_path) and img_path.endswith(('.png', '.jpg', '.jpeg')):
            img = cv2.imread(img_path)
            if img is not None:
                images.append(img)
    return images

def main(root_dir):
    for dataset_type in ['train', 'test']:
        dataset_path = os.path.join(root_dir, dataset_type)
        if os.path.isdir(dataset_path):
            for folder_name in sorted(os.listdir(dataset_path)):
                folder_path = os.path.join(dataset_path, folder_name)
                if os.path.isdir(folder_path):
                    print(f'Reading images from folder: {folder_path}')
                    images = read_images_from_folder(folder_path)
                    num_images = len(images)
                    print(f'Number of images in {folder_path}: {num_images}')

if __name__ == "__main__":
    root_directory = 'dataset'
    main(root_directory)
```
