```python
import os
import cv2
import json
import random
import matplotlib.pyplot as plt
```

```python
def get_random_image_filenames(folder_path, num_files=10):
    image_files = [f for f in sorted(os.listdir(folder_path)) if f.endswith(('.png', '.jpg', '.jpeg'))]
    return random.sample(image_files, min(num_files, len(image_files)))
```

```python
def load_json_for_images(folder_path, image_filenames):
    json_data = {}
    for filename in image_filenames:
        json_filename = filename.replace('.jpg', '.json').replace('.jpeg', '.json').replace('.png', '.json')
        json_path = os.path.join(folder_path, json_filename)
        if os.path.isfile(json_path):
            try:
                with open(json_path, 'r', encoding='utf-8') as f:
                    if os.path.getsize(json_path) > 0:  # Check if the file is not empty
                        data = json.load(f)
                        json_data[json_filename] = data
            except (UnicodeDecodeError, json.JSONDecodeError) as e:
                print(f"Error reading {json_path}: {e}")
    return json_data
```

```python
def load_images(folder_path, image_filenames):
    images = {}
    for filename in image_filenames:
        img_path = os.path.join(folder_path, filename)
        if os.path.isfile(img_path):
            img = cv2.imread(img_path)
            if img is not None:
                images[filename] = img
    return images
```

```python
def display_images_with_bounding_boxes(images, json_data):
    for filename, img in images.items():
        json_filename = filename.replace('.jpg', '.json').replace('.jpeg', '.json').replace('.png', '.json')
        if json_filename in json_data:
            data = json_data[json_filename]
            regions = data.get('regions', [])
            for region in regions:
                shape_attributes = region.get('shape_attributes', {})
                x = shape_attributes.get('x', 0)
                y = shape_attributes.get('y', 0)
                width = shape_attributes.get('width', 0)
                height = shape_attributes.get('height', 0)
                cv2.rectangle(img, (x, y), (x + width, y + height), (0, 255, 0), 2)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()
```

```python
def main(root_dir, num_images=10):
    all_images = {}
    all_json_data = {}
    
    for dataset_type in ['train', 'test']:
        dataset_path = os.path.join(root_dir, dataset_type)
        if os.path.isdir(dataset_path):
            for folder_name in sorted(os.listdir(dataset_path)):
                folder_path = os.path.join(dataset_path, folder_name)
                if os.path.isdir(folder_path):
                    image_filenames = get_random_image_filenames(folder_path, num_files=num_images)
                    images = load_images(folder_path, image_filenames)
                    json_data = load_json_for_images(folder_path, image_filenames)
                    all_images.update(images)
                    all_json_data.update(json_data)
                    
                    # If we have enough images, stop
                    if len(all_images) >= num_images:
                        break
            if len(all_images) >= num_images:
                break
    
    return all_images, all_json_data
```

```python
if __name__ == "__main__":
    root_directory = 'dataset'
    num_images = 10
    images, json_data = main(root_directory, num_images=num_images)
    display_images_with_bounding_boxes(images, json_data)
```
