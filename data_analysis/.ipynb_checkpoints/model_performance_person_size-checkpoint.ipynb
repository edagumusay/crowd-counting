```python
import os
import numpy as np
import cv2
from ultralytics import YOLO
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
model = YOLO('../trained_models/best.pt')

test_images_path = '../dataset_size64/test/images'
test_labels_path = '../dataset_size64/test/labels'
```

# Approach 1: Using Intersection over Union (IoU)

- The Bounding boxes are classified as small, medium, or large based on pixel area of the bounding boxes.
- The IoU between each ground truth box and the predicted box is calculated for each category.
- IoU score distributions are visualized for each size category to analyze model performance across different sizes of persons in an image.

```python
# Define size thresholds (in pixels) for categories
size_thresholds = {
        'small': (0, 32*32),
        'medium': (32*32, 96*96),
        'large': (96*96, float('inf'))
    }
```

```python
iou_scores = {'small': [], 'medium': [], 'large': []}
```

```python
def load_label(file_path):
    boxes = []
    with open(file_path, 'r') as f:
        for line in f.readlines():
            values = list(map(float, line.strip().split()))
            class_id, x_center, y_center, width, height = values[0], values[1], values[2], values[3], values[4]
            boxes.append((x_center, y_center, width, height))
    return boxes

def convert_bbox_to_pixels(bbox, image_shape):
    # Convert YOLO format to pixel coordinates and calculate area
    x_center, y_center, width, height = bbox
    img_h, img_w = image_shape[:2]
    x1 = int((x_center - width / 2) * img_w)
    y1 = int((y_center - height / 2) * img_h)
    x2 = int((x_center + width / 2) * img_w)
    y2 = int((y_center + height / 2) * img_h)
    area = (x2 - x1) * (y2 - y1)
    return (x1, y1, x2, y2), area

def categorize_bbox_size(area):
    for category, (min_size, max_size) in size_thresholds.items():
        if min_size <= area < max_size:
            return category
    return 'unknown'

def calculate_iou(box1, box2):
    x1, y1, x2, y2 = box1
    x1_p, y1_p, x2_p, y2_p = box2
    
    xi1 = max(x1, x1_p)
    yi1 = max(y1, y1_p)
    xi2 = min(x2, x2_p)
    yi2 = min(y2, y2_p)
    
    inter_area = max(0, xi2 - xi1) * max(0, yi2 - yi1)
    box1_area = (x2 - x1) * (y2 - y1)
    box2_area = (x2_p - x1_p) * (y2_p - y1_p)
    
    union_area = box1_area + box2_area - inter_area
    iou = inter_area / union_area if union_area > 0 else 0
    return iou
```

```python
for img_file in os.listdir(test_images_path):
    img_path = os.path.join(test_images_path, img_file)
    label_path = os.path.join(test_labels_path, os.path.splitext(img_file)[0] + '.txt')
    
    # Load image and labels
    image = cv2.imread(img_path)
    image_shape = image.shape
    true_boxes = load_label(label_path)
    
    results = model.predict(img_path, verbose=False)[0]
    
    for true_bbox in true_boxes:
        true_box_pixels, true_area = convert_bbox_to_pixels(true_bbox, image_shape)
        category = categorize_bbox_size(true_area)
        
        best_iou = 0
        for pred_box in results.boxes.xyxy:
            pred_box = tuple(map(int, pred_box[:4].tolist()))
            iou = calculate_iou(true_box_pixels, pred_box)
            best_iou = max(best_iou, iou)
        
        # store IoU by category
        if category in iou_scores:
            iou_scores[category].append(best_iou)
```

## Results visualization

```python
 fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for ax, (category, ious) in zip(axes, iou_scores.items()):
    ax.hist(ious, bins=20, density=True, alpha=0.7, color='blue')
    ax.set_title(f'Relative IoU Distribution - {category} bboxes')
    ax.set_xlabel('IoU')
    ax.set_ylabel('Relative Frequency')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

- For small bounding boxes, there is a high concentration near 0, indicating that the model struggles to achieve high overlap with the ground truth boxes
- On medium boxes, the model performs significantly better than small ones.
- For the large bounding boxes, the model seems to performs very well, with distribution similar to the medium category but slightly skewed towards higher IoU values reaching around 0.9

# Approach 2: Based on Predicted vs. Ground Truth Bounding Box Counts

- Here, images themselves are categorezed based on the dominant size of detected bounding boxes. Categories are similar to those of approach one: small, medium and large.
- For each image, we count the number of predicted bounding boxes from the model and the number of ground truth boxes in the labels.
- Using binning, the confusion matrices are generated for each category to show the relationship between predicted and ground truth counts
- Additionally, weighted precision, recall, and F1 scores are computed for each size category.

```python
def load_ground_truth_count(label_path):
    with open(label_path, 'r') as file:
        lines = file.readlines()
        return len(lines)

def calculate_bbox_size(bbox):
    x1, y1, x2, y2 = bbox
    return (x2 - x1) * (y2 - y1)

def categorize_sizes(bbox_sizes):
    size_counts = {'small': 0, 'medium': 0, 'large': 0}
    for size in bbox_sizes:
        for category, (min_size, max_size) in size_thresholds.items():
            if min_size <= size < max_size:
                size_counts[category] += 1
                break
    return max(size_counts, key=size_counts.get)  # Return the dominant size category


def evaluate_model_size_distribution(model, images_path, labels_path):
    image_categories = []
    ground_truth_counts = []
    predicted_counts = []

    for image_file in os.listdir(images_path):
        image_path = os.path.join(images_path, image_file)
        image = cv2.imread(image_path)
        results = model.predict(image, verbose=False)

        # Calculate the size of each detected bounding box
        bbox_sizes = [calculate_bbox_size(box.xyxy[0].cpu().numpy()) for box in results[0].boxes]
        
        # Categorize image based on the dominant size category
        image_category = categorize_sizes(bbox_sizes)
        image_categories.append(image_category)

        # Count the number of detections and ground truths
        predicted_count = len(results[0].boxes)
        label_file = image_file.replace('.jpg', '.txt')
        label_path = os.path.join(labels_path, label_file)
        ground_truth_count = load_ground_truth_count(label_path)

        predicted_counts.append(predicted_count)
        ground_truth_counts.append(ground_truth_count)

    return image_categories, ground_truth_counts, predicted_counts


def create_size_distribution_confusion_matrix(image_categories, ground_truth_counts, predicted_counts):
    bins = [0, 5, 10, 15, 20, 30, 50, 100, 200, 300]
    labels = [f'{bins[i]}-{bins[i+1]}' for i in range(len(bins)-1)]

    predicted_binned = pd.cut(predicted_counts, bins=bins, labels=labels, right=False)
    ground_truth_binned = pd.cut(ground_truth_counts, bins=bins, labels=labels, right=False)


    # Create DataFrame for analysis
    data = pd.DataFrame({
        'Category': image_categories,
        'Ground Truth Count': ground_truth_binned,
        'Predicted Count': predicted_binned
    })

    # Create a confusion matrix grouped by size category
    confusion_matrices = {}
    for category in size_thresholds.keys():
        subset = data[data['Category'] == category]
        cm = confusion_matrix(subset['Ground Truth Count'], subset['Predicted Count'], labels=labels)
        confusion_matrices[category] = cm

    return confusion_matrices, labels, data
```

```python
# Evaluate model and categorize images by dominant size category
image_categories, ground_truth_counts, predicted_counts = evaluate_model_size_distribution(
    model, test_images_path, test_labels_path
)
```

```python
# Create confusion matrices for each size category
confusion_matrices, labels, data = create_size_distribution_confusion_matrix(
    image_categories, ground_truth_counts, predicted_counts
)
```

```python
def calculate_weighted_precision_recall(cm):
    precisions = []
    recalls = []
    supports = []

    num_classes = len(cm)
    for i in range(num_classes):
        tp = cm[i, i]
        fp = cm[:, i].sum() - tp
        fn = cm[i, :].sum() - tp
        support = cm[i, :].sum()  # the total number of true instances of class i

        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0

        precisions.append(precision)
        recalls.append(recall)
        supports.append(support)

    # weighted average precision and recall
    total_support = sum(supports)
    weighted_precision = sum(p * s for p, s in zip(precisions, supports)) / total_support
    weighted_recall = sum(r * s for r, s in zip(recalls, supports)) / total_support

    return weighted_precision, weighted_recall
```

## Category small

```python
category = 'small'
cm = confusion_matrices[category]
ig, ax = plt.subplots(figsize=(10, 8))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap='Blues', ax=ax, values_format='d')
plt.title(f"Confusion Matrix, category - {category}")
plt.xticks(rotation=45)
plt.show()


# Metrics calculation
weighted_precision, weighted_recall = calculate_weighted_precision_recall(cm)
weighted_f1 = 2 * (weighted_precision * weighted_recall) / (weighted_precision + weighted_recall) if (weighted_precision + weighted_recall) > 0 else 0

metrics_df = pd.DataFrame({
    "Metric": ["Weighted Precision", "Weighted Recall", "Weighted F1 Score"],
    "Value": [f"{weighted_precision:.2f}", f"{weighted_recall:.2f}", f"{weighted_f1:.2f}"]
})

metrics_df.style.set_properties(**{'text-align': 'center'}).set_table_styles(
    [{'selector': 'th', 'props': [('text-align', 'center')]}]
)
```

# Summary

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

metrics_dict = {
    "Metric": ["Weighted Precision", "Weighted Recall", "Weighted F1 Score"]
}
for i, category in enumerate(size_thresholds.keys()):
    cm = confusion_matrices[category]
    ax = axes[i] 
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap='Blues', ax=ax, values_format='d')
    ax.set_title(f"category - {category}")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

    # Print weighted precision and recall for each category
    weighted_precision, weighted_recall = calculate_weighted_precision_recall(cm)
    weighted_f1 = 2 * (weighted_precision * weighted_recall) / (weighted_precision + weighted_recall) if (weighted_precision + weighted_recall) > 0 else 0
    
    metrics_dict[category.capitalize()] = [
        f"{weighted_precision:.2f}",
        f"{weighted_recall:.2f}",
        f"{weighted_f1:.2f}"
    ]

plt.tight_layout()
plt.show()

# Display the table
metrics_df = pd.DataFrame(metrics_dict)
metrics_df.style.set_properties(**{'text-align': 'center'}).set_table_styles(
    [{'selector': 'th', 'props': [('text-align', 'center')]}]
)
```

- The model performs best on large persons, with moderate success on medium objects, and struggles the most with small objects.
- The models detection accuracy is also positively correlated with person size in an image
